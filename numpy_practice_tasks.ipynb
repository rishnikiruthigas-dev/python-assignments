```python
#Question:1 Find its transpose.

np.transpose(a)
```

```python
#Create a 3 × 3 matrix.

a=np.arange(2,11)
print(a)
```

```python
x=np.reshape(a,(3,3))
print(x)
```

```python
#Question:2 Find the sum of each row

print(x)
print("row 1 sum:",x[0:1].sum())
print("row 2 sum:",x[1:2].sum())
print("row 3 sum:",x[2:3].sum())
```

```python
#Question:3 Find the sum of each column

print(x)
print("column 1 sum:",x[:,0].sum())
print("column 2 sum:",x[:,1].sum())
print("column 3 sum:",x[:,2].sum())
```
