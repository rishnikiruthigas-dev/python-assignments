```python
import numpy as np
students = np.array(["Arun", "Priya", "Kavin", "Divya", "Rahul"])
marks = np.array([450, 480, 390, 420, 350])

print("Students:", students)
print("Marks:", marks)
print("Total Marks =", np.sum(marks))
print("Average Marks =", np.mean(marks))
print("Highest Mark =", np.max(marks))
print("Lowest Mark =", np.min(marks))
```

```python
import pandas as pd
data = {"Name": ["Arun", "Priya", "Kavin", "Divya", "Rahul","ajay"],"Marks": [450, 480, 390, 420, 350,460]}
df = pd.DataFrame(data)

print("First 5 Records:")
print(df.head())
print("\nLast 5 Records:")
print(df.tail())
print("\nSummary:")
print(df.describe())
print("\nHighest Mark:")
print(df["Marks"].max())
print("\nAverage Mark:")
print(df["Marks"].mean())
```

```python
import matplotlib.pyplot as plt
students = np.array(["Arun", "Priya", "Kavin", "Divya", "Rahul"])
marks = np.array([450, 480, 390, 420, 350])
plt.scatter(students,marks)
plt.bar(students,marks)
plt.title("students marks")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = { "Name": ["Arun", "Priya", "Kavin", "Divya", "Rahul"], "Marks": [450, 480, 390, 420, 350]}

df = pd.DataFrame(data)
sns.barplot(x="Name", y="Marks", data=df)
plt.title("Bar Plot")
plt.show()
```

```python
sns.lineplot(x="Name", y="Marks", data=df)
plt.title("Line Plot")
plt.show()
```

```python
sns.scatterplot(x="Name", y="Marks", data=df)
plt.title("Scatter Plot")
plt.show()
```

```python
sns.histplot(df["Marks"])
plt.title("Histogram")
plt.show()
```

```python
sns.boxplot(y="Marks", data=df)
plt.title("Box Plot")
plt.show()
```
