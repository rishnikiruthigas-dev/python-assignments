```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sample_data.csv")
```

```python
#column rename:
df=df.rename(columns={"Years Exp":"Years of Experience","Emp ID":"Employee ID","Salary (INR)":"Salary"})
df
```

```python
#  remove duplicates
df.drop_duplicates()
df
```

```python
# Handle null values (drop) 
df.dropna()
```

```python
# Handle null values (fill)
df.fillna({"Employee ID": 000,"Employee Name": "abcd","AGE": 20,"Remarks": "good"}, inplace=True)
df
```

```python
#Change data types 
print(df.dtypes)
```

```python
print(df["AGE"].unique())
```

```python
print(df["AGE"].dtype)
print(df["AGE"].head(10))
```

```python
print(df["Years of Experience"].unique())
```

```python
df["Years of Experience"] = df["Years of Experience"].replace("three", 3)
df["Years of Experience"] = pd.to_numeric(df["Years of Experience"],errors="coerce")
df["Years of Experience"] = df["Years of Experience"].fillna(df["Years of Experience"].median())
df["Years of Experience"] = df["Years of Experience"].astype(int)
df.head()
```

```python
print(df.dtypes)
```

```python
df.rename(columns={"AGE": "Age"}, inplace=True)
```

```python
df.columns
```

```python
#Standardize names 

df.columns = df.columns.str.strip()
df["Employee Name"] = df["Employee Name"].str.title()
df["Employee Name"].head()
```

```python
#Standardize skills 

df["Skill Set"] = df["Skill Set"].astype(str).str.title()
df["Skill Set"].head()
```

```python
#Clean salary values 

print(df["Salary"].unique())
```

```python
df["Salary"] = df["Salary"].replace({"5,50,000": "550000","650000 INR": "650000","$500000": "500000",
    "4.8L": "480000","Nine Lakhs": "900000","99999999": "990000",0: "500000"})
df
```

```python
#Parse dates 

print(df["Date Of Joining"].dtype)
```

```python
print(df["Salary"].dtype)
```

```python


df["Salary"]=pd.to_numeric(df["Salary"])
```

```python
#Detect and remove outliers 
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1
df = df[(df["Salary"] >= Q1 - 1.5 * IQR) & (df["Salary"] <= Q3 + 1.5 * IQR)]
df.shape
print("Rows after removing outliers:", df.shape[0])
```

```python
df.info()
```

```python
df["Age"]=df["Age"].str.replace("twenty six","26")

df["Age"]=pd.to_numeric(df["Age"])
```

```python
df["Years of Experience"]=pd.to_numeric(df["Years of Experience"])
```

```python
#Validate ages and experience 

df["Age"]=pd.to_numeric(df["Age"])
df = df[(df["Age"] >= 18) & (df["Years of Experience"] >= 0) & (df["Years of Experience"] <= df["Age"])]
print(df)
```

```python
#Encode categorical variables 
df = pd.get_dummies(df, columns=["Remarks"])
df.head()
```

```python
#Scale numerical columns 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[["Age", "Years of Experience", "Salary"]] = scaler.fit_transform(df[["Age", "Years of Experience", "Salary"]])
df.head()
```

```python
#Feature selection
df = df.drop(["Employee ID", "Date Of Joining"], axis=1, errors="ignore")
df.head()
```
